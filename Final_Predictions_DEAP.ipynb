```
all_eeg_data = pd.DataFrame(index=range(0),columns=range(0,39))

for m in range(1,33):
    fl = f'/content/drive/My Drive/data_preprocessed_python/s{m}.dat'
    with open(fl, 'rb') as fl: content = pickle.load(fl, encoding='latin1')

    data_ml=[]
    label_ml=[]

    for i in range(0,40):
      data_ml.append(content["data"][i])
      label_ml.append(content["labels"][i])

    np.save('data_ml', np.array(data_ml), allow_pickle=True, fix_imports=True)
    np.save('label_ml', np.array(label_ml), allow_pickle=True, fix_imports=True)

    #df_data_label=pd.DataFrame(list(label_ml))

    dt1=[]

    for i in range(0,40):
      for j in range(0,40):
        dt1.append(mean(data_ml[i][j]))

    df_data_eeg=pd.DataFrame(list(dt1))

    x=0
    mean_eeg=[]
    for i in range(0,40):
      mean_eeg.append(df_data_eeg.iloc[x:x+40,:].values)
      x=x+40

    x=np.reshape(mean_eeg,(40,40))

    mean_eeg_df_final = pd.DataFrame(data=x)

    print(m)

    all_eeg_data=pd.concat([all_eeg_data,mean_eeg_df_final],ignore_index=True)

all_eeg_data.shape


```

```
#valence,arousal,dominance,liking
all_valence_data = pd.DataFrame(index=range(0),columns=range(1))
all_arousal_data = pd.DataFrame(index=range(0),columns=range(1))
all_dominance_data = pd.DataFrame(index=range(0),columns=range(1))
all_liking_data = pd.DataFrame(index=range(0),columns=range(1))

for m in range(1,33):
    fl = f'/content/drive/My Drive/data_preprocessed_python/s{m}.dat'
    with open(fl, 'rb') as fl: content = pickle.load(fl, encoding='latin1')
    
    label_ml=[]
    
    for i in range(0,40):
      label_ml.append(content["labels"][i])

    np.save('label_ml', np.array(label_ml), allow_pickle=True, fix_imports=True)
    df_data_label=pd.DataFrame(list(label_ml))

    y_val=df_data_label.iloc[:,0].values
    y_ar=df_data_label.iloc[:,1].values
    y_dom=df_data_label.iloc[:,2].values
    y_lk=df_data_label.iloc[:,3].values
  

    y_ar_list=[]
    y_dom_list=[]
    y_lk_list=[]
    y_val_list=[]


    for i in range(0,40):
      y_val_list.append(int(y_val[i]//5))
      y_ar_list.append(int(y_ar[i]//5))
      y_dom_list.append(int(y_dom[i]//5))
      y_lk_list.append(int(y_lk[i]//5))

    y_val_df=pd.DataFrame(list(y_val_list))
    y_ar_df=pd.DataFrame(list(y_ar_list))
    y_dom_df=pd.DataFrame(list(y_dom_list))
    y_lk_df=pd.DataFrame(list(y_lk_list))
    
    #print(y.shape)
    
    all_valence_data=pd.concat([all_valence_data,y_val_df],ignore_index=True)
    all_arousal_data=pd.concat([all_arousal_data,y_ar_df],ignore_index=True)
    all_dominance_data=pd.concat([all_dominance_data,y_dom_df],ignore_index=True)
    all_liking_data=pd.concat([all_liking_data,y_lk_df],ignore_index=True)

print(all_arousal_data.shape)
print(all_valence_data.shape)
print(all_dominance_data.shape)
print(all_liking_data.shape)
```

```
gaus_predictions=[]
dt_predictions=[]
rf_predictions=[]
knn_predictions=[]
features=[all_valence_data, all_arousal_data, all_dominance_data, all_liking_data]
f_names=['valence','arousal','dominance','liking']
for i in range(0,4):
  y=features[i].values
  y=y.astype('int')
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
  #GaussianNB()
  nbclassifier=GaussianNB()
  nbclassifier.fit(X_train, y_train)
  nb_y_pred = nbclassifier.predict(X_test)
  print("GaussianNB for ", f_names[i])
  gaus_predictions.append(accuracy_score(y_test, nb_y_pred)*100)
  print("Test Accuracy: {:.2f}%".format(accuracy_score(y_test, nb_y_pred)*100))
  print(classification_report(y_test, nb_y_pred))

  #DecisionTreeClassifier
  from sklearn.tree import DecisionTreeClassifier
  dtclassifier= DecisionTreeClassifier(criterion='entropy')
  dtclassifier.fit(X_train,y_train)
  # Make predictions using the testing set
  y_pred_DT = dtclassifier.predict(X_test)
  print("DecisionTreeClassifier for ", f_names[i])
  dt_predictions.append(accuracy_score(y_test, y_pred_DT)*100)
  print(" Test Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_DT)*100))
  print(classification_report(y_test, y_pred_DT))

  #RandomForestClassifier
  from sklearn.ensemble  import RandomForestClassifier
  RF = RandomForestClassifier(n_estimators=100, criterion='entropy', random_state=0)  
  RF.fit(X_train, y_train)
  y_pred_RF = RF.predict(X_test)
  print("RandomForestClassifier for ", f_names[i])
  rf_predictions.append(accuracy_score(y_test, y_pred_RF)*100)
  print("Test Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_RF)*100))
  print(classification_report(y_test, y_pred_RF))

  #KNN
  knn = KNeighborsClassifier(n_neighbors =9)  # n_neighbors means k
  knn.fit(X_train, y_train)
  y_pred_knn = knn.predict(X_test)
  print("KNNClassifier for ", f_names[i])
  knn_predictions.append(accuracy_score(y_test, y_pred_knn)*100)
  print("Test Accuracy: {:.2f}%".format(accuracy_score(y_test, y_pred_knn)*100))
  print(classification_report(y_test, y_pred_knn))
```

```
df_predictions = pd.DataFrame(list(zip(gaus_predictions,dt_predictions,rf_predictions,knn_predictions)), 
               columns =['Gaussian', 'DT','RF','KNN'])
axis = df_predictions[['Gaussian', 'DT','RF','KNN']].plot(kind="bar",figsize=(20,5),xlabel='Models',ylabel='Accuracy')
fig = axis.get_figure()
```
